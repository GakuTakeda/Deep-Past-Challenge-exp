# Print CSV heads in data/

Show header and first row for each CSV file in `data/`.

```python
from pathlib import Path
import csv

data_dir = Path('..') / 'data'

csv_files = sorted(
    path for path in data_dir.glob('*.csv')
    if ':' not in path.name
)

if not csv_files:
    print('No CSV files found.')
else:
    for csv_path in csv_files:
        print(f'=== {csv_path.name} ===')
        with csv_path.open('r', encoding='utf-8', newline='') as f:
            reader = csv.reader(f)
            header = next(reader, None)
            first_row = next(reader, None)

        if header is None:
            print('(empty file)')
        else:
            print('header:', header)
            print('first_row:', first_row if first_row is not None else '(no data rows)')
        print()
```
